As we can see, the datatype of the `Date` column is `str` or string

We will now convert it into a `datetime` format

We are doing this so we can extract more information out of it such as year and month

```python
df["Date"] = pd.to_datetime(df["Date"])
```

Now we will extract the year, month, day and weekday from the `Date` column

We will put this new information into their new respective columns

```python
df["Date"].head()
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.day_name()
```

In our dataset we have 2 columns which are `Price` and `Discount`

`Price` = this tells the price of the product

`Discount` = this tells the discount % applied on the product

```python
df[["Price", "Discount"]].head()
```

What we'll do here is create a new column named `Discounted Price`

This will contain the `price - discount` as the new price

```python
df["DiscountedPrice"] = df["Price"] * (1 - df["Discount"] / 100)
```

```python
df[["Price", "Discount", "DiscountedPrice"]].head()
```

In our dataset regarding the units sold, we have 2 important columns

`Units Sold` = this tells the no. of units sold

`Inventory Level` = this tells the total no. of units we had

```python
df[["Units Sold", "Inventory Level"]].head()
```

From this we will extract a new column called `Sell-Through Rate`

What it tells is the % of units sold from the entire inventory we had

This helps us understand the demand of that specific product

```python
df["SellThroughRate"] = df["Units Sold"] / df["Inventory Level"]
```

```python
df[["Units Sold", "Inventory Level", "SellThroughRate"]].head()
```

Now that we have made a couple of changes in our dataset let us look the entire dataset 

```python
df.head()
```

Now let us check the different categories in our dataset along with the different stats about their demands

This will help us know which categories have most demands 

```python
df.groupby("Category")["Demand"].agg(["mean", "median", "std"]).sort_values("mean", ascending=False)
```

Now let us check the demand based on the region and season and see how different geographical features affect the demand

```python
df.groupby(["Region", "Seasonality"])["Demand"].mean()
```

**Highest:** Summer (in all regions)

**Lowest:** Spring (in all regions)

Now lets check the demand of each category month wise to help us understand which categories have the most demand in which months

```python
pd.pivot_table(df, values="Demand", index="Month", columns="Category", aggfunc="mean")
```

Similarly, we will check the category wise demand in different regions as well

```python
pd.pivot_table(df, values="Demand", index="Region", columns="Category", aggfunc="mean")
```

We can see that in the west region there is no demand for `Toys`

We should also cross-check this by further checking the dataset for demand of `Toys` in `West` region

```python
df[(df["Region"] == "West") & (df["Category"] == "Toys")]
```
